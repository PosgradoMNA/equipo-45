######Lo malo es que, si falta un dato, la columna entera se borra, en este caso, se borraron columnas enteras

###### Mejor limpiaremos con el siguiente método, sustituiremos los valores nulos con el promedio de los valores en la columna, pero a partir de la columna X6 porque a partir de esa columna contamos con valores de peso.


```python
#Pasamos el archivo csv a un dataframe
PathData = "/content/cc.csv"
df2 = pd.read_csv(PathData, index_col=0) 
```

```python
ndf = df2.copy()
ndf
```

```python
#Pongamos las columnas en un arreglo para irlas recorriendo posteriormente
columnas = ndf.columns
columnas
```

```python
#No vamos a tomar las primeras columnas (Amount of the given credit, gender, education, Marital status, Age) 
#ya que son variables que perfilan al cliente, pero no consideramos su valor como un valor de peso, si no como un valor categórico.
columnas = columnas[5:]
columnas
```

```python
#Por cada columna en la que habían datos nulos, éstos se sustituyeron por la media de esa columna.
for col in columnas:
  ndf[col].fillna(value = ndf[col].mean(), inplace = True)
ndf
```

```python
#Al final, seguirán existiendo valores nulos, pero menos y más acotados
ndf.isnull().values.any()
```
